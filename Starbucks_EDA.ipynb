### Questions and Data Visualizations


```python
df['Beverage'].unique()
```

### Which Starbucks drink has the highest calories from the dataset?¶


```python
df.calories = df[['Beverage','Calories']].sort_values('Calories',ascending=False)

df.calories
```

```python
#plot relationship between calories and beverages
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.countplot(y='Beverage', data=df.calories)
```

##### A: White Chocolate Mocha (Without Whipped Cream) is our winner!

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y="Calories", data=df)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Calories")
plt.show()
```

### Which drink has the highest fat content?

```python
df.fat = df[['Beverage',' Total Fat (g)']].sort_values(' Total Fat (g)',ascending=False)

df.fat
```

##### A: Hot Chocolate (Without Whipping Cream)

### Which drink has the highest protein?

```python
df.protein = df[['Beverage',' Protein (g) ']].sort_values(' Protein (g) ',ascending=False)

df.protein
```

##### A: Banana Chocolate Smoothie is highest in protein

### Q2. Which drink has the highest calories from the Starbucks classic espresso drinks?

```python
#extract classic espress drinks from Beverage_category
classic = df.loc[(df['Beverage_category'] == 'Classic Espresso Drinks')]
classic.head()
```

```python
# data visualization to see which classic espresso drink has the highest calories
sns.barplot(x="Beverage", y="Calories", data=classic)
plt.xticks(rotation=30, ha='right')
plt.title("Starbucks Classic Drinks Calories")
```

##### A: Caffe Mocha (without whipped cream) has the highest kcal, followed by vanilla latte(or other flavoured latte) from the Starbucks classic espresso drinks.

### Q3. Highest Sugar Drink at Starbucks?

```python
#To see which Starbucks drink has the highest sugars
df.sugar = df[['Beverage'," Sugars (g)"]].sort_values(" Sugars (g)",ascending=False)

df.sugar
```

```python
# data visualization to see the which Starbucks has the highest calories
plt.figure(figsize=(12, 8))
sns.barplot(x="Beverage", y="Calories", data=df)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Calories")
plt.show()
plt.savefig("Calories.jpg")
```

##### A: Java Chip (without Whipped Cream) is highest in calories.

### Q4. Highest Sugar Drink from Signature Espresso Drinks?¶


```python
signature = df.loc[(df['Beverage_category'] == 'Signature Espresso Drinks')]
signature.sort_values(' Sugars (g)', ascending=False).head()
```

```python
sns.barplot(x='Beverage', y=' Sugars (g)', data=signature)
plt.xticks(rotation=30, ha='right')
plt.title("Starbucks Signature Espresso Drinks Sugars(g)")
plt.savefig("Signature Espresso Sugars.jpg", dpi=(300), bbox_inches='tight')
```

##### A:Caramel Apple Spice (without whipped cream), followed by White chocolate mocha (whithout whipped cream) have the highest sugars

### Now we want to explore correlation between calories and sugar content in all drinks.

```python

sug_cal = sns.relplot(data=df,kind='line', x=" Sugars (g)", y="Calories")


plt.savefig("sugar_cal_corr.jpg", dpi=(300), bbox_inches='tight')
```

##### The higher the sugar the higher the calories.

### Map the correlation between different variables

```python
#Get correlation between different variables 
corr = df.corr(method='kendall')
plt.figure(figsize=(15,8))
heatmap = sns.heatmap(corr, annot=True, cmap="Greens")

fig=heatmap.get_figure()

fig.savefig("Heatmap.jpg", dpi=(300), bbox_inches='tight')
```

##### We can see that calories have highest correlation with:
1.Cholesterol, followed by 

2.sugars

3.total carbs

4.fat

5.protein

```python
 macros= df.groupby('Beverage')[[" Sugars (g)", 'Trans Fat (g) ', " Protein (g) "]].agg(['mean']).plot.bar(figsize=(12,8))
```

```python
fig=macros.get_figure()

fig.savefig("Beverage Macro Ratio.jpg", dpi=(300), bbox_inches='tight')
```

Here we can see the distribution of major macros inside of all drinks. It is obvious that sugar is the main ingredient in most of them.
